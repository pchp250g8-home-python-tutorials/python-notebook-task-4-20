Шестизначное число начинается с цифры 2. 
Если 2 переставить с первого места на последнее, 
то получится число в 3 раза больше изначального. 
Найти начальное число и в ответ  записать сумму его цифр. 

```python
from sympy.abc import x
from sympy import solve
from sympy import Eq

s = 0
res = solve(Eq(10 * x + 2, 3 * (200000 + x)), x)
ans = 200000 + int(res.pop())
iNum = ans
while (iNum > 0):
    s += iNum % 10
    iNum //= 10
print("Number:", ans)
print("The sum of digits:", s)
```
